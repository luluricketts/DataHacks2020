```python
def makeNotNull(column):    
    print('Originally', column.isna().sum(), 'Null values')
    
    mean = column.mean()
    print('Original mean is', mean)
    
    print('Filling NaN with mean')
    column.fillna(value=mean, inplace=True)
    print('There are now', column.isna().sum(), 'null values')
    
    new_mean = column.mean()
    print('New mean after filling NaN:', new_mean)
    print('Difference is', abs(mean-new_mean), '\n')
    
```

```python
makeNotNull(df_barts_hotspots['Daily Mean Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['Daily Range - Lower Bound Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['Daily Range - Upper Bound Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['AM Mean Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['AM Range - Lower Bound Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['AM Range - Upper Bound Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['PM Mean Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['PM Range - Lower Bound Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['PM Range - Upper Bound Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['Midday Mean Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['Midday Range - Lower Bound Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['Midday Range - Upper Bound Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['Evening Mean Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['Evening Range - Lower Bound Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['Evening Range - Upper Bound Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['Early Morning Mean Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['Early Morning Range - Lower Bound Travel Time (Seconds)'])
makeNotNull(df_barts_hotspots['Early Morning Range - Upper Bound Travel Time (Seconds)'])
```
